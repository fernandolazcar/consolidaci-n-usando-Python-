hay que ver todas las pestañas que tiene el archivo en el que queremos trabajar en este caso usaremos este codigo para saber que nombres tienen

```python
import openpyxl

# Nombre del archivo Excel
archivo_excel = "/content/supermarket_sales - Sheet1.xlsx"

# Cargar el libro de trabajo (workbook)
libro_trabajo = openpyxl.load_workbook(archivo_excel)

# Obtener los nombres de las pestañas
nombres_de_pestanas = libro_trabajo.sheetnames

# Mostrar los nombres de las pestañas
for nombre in nombres_de_pestanas:
    print(nombre)

# Cerrar el libro de trabajo
libro_trabajo.close()
```
